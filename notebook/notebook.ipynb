```python
import pandas as pd
import spacy
import numpy as np
```

```python
real_news = pd.read_csv("../data/True.csv")
real_news
```

```python
fake_news = pd.read_csv("../data/Fake.csv")
fake_news
```

```python
fake_news['label'] = 0
```

```python
real_news['label'] = 1
```

```python
news_df = pd.concat([real_news[0:10000],fake_news[0:10000]], axis=0)
```

```python
news_df.tail()
```

```python
news_df.drop(['title','subject','date'], axis=1, inplace=True)
```

```python
nlp = spacy.load('en_core_web_lg')
```

```python
def preprocess(text):
    doc = nlp(text)
    filtered_tokens = []
    for token in doc:
        if token.is_stop or token.is_punct:
            continue
        filtered_tokens.append(token.lemma_)
    return " ".join(filtered_tokens)
```

```python
news_df['processed_text'] = news_df['text'].apply(preprocess)
```

```python
news_df['vector'] = news_df['processed_text'].apply(lambda x: nlp(x).vector)
```

```python
news_df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

input = news_df.vector
output = news_df.label
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    input,
    output,
    test_size=0.3,
    random_state=43,
    shuffle=True,
    stratify=output
    )
```

```python
X_train = np.stack(X_train)
X_test = np.stack(X_test)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier

MNB_classifier = MultinomialNB()
MNB_classifier.fit(X_train_scaled,y_train)
```

```python
LR_classifier = LogisticRegression()
LR_classifier.fit(X_train,y_train)
```

```python
KNN_classifier = KNeighborsClassifier(n_neighbors= 5, metric='euclidean')
KNN_classifier.fit(X_train_scaled,y_train)
```

```python
y_pred_NB = MNB_classifier.predict(X_test_scaled)
y_pred_LR = LR_classifier.predict(X_test)
y_pred_KNN = KNN_classifier.predict(X_test_scaled)
```

```python
print("Logistic Regression:- \n",classification_report(y_test,y_pred_LR))
print("Multinomial Naive Bayes:- \n",classification_report(y_test,y_pred_NB))
print("K Nearest Neighbors:- \n",classification_report(y_test,y_pred_KNN))
```
